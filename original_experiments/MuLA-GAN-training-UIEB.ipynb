```python
%%writefile PSNR-SSIM-UCIQE-UIQM-Python/evaluate.py
# -- coding: utf-8 --

import numpy
import math
from PIL import Image
import numpy as np
import glob, os
import scipy.ndimage
import cv2

# --- CORRECTED IMPORTS ---
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
# -------------------------

# --- UIQM HELPER FUNCTIONS (More Robust Error Handling) ---
def uiconm(img):
    try:
        # Check if image is grayscale, convert to color if so
        if len(img.shape) == 2 or img.shape[2] == 1:
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

        if img is None or img.size == 0: return 0 # Handle empty image

        img = img.astype(np.float32)/255.0
        # Ensure image has 3 channels after potential conversion
        if img.shape[2] != 3: return 0

        img_r = img[:,:,0]
        img_g = img[:,:,1]
        img_b = img[:,:,2]
        alpha_r = 0.1
        alpha_g = 0.1
        alpha_b = 1 - alpha_r - alpha_g

        def calculate_contrast(channel):
            # Use nanpercentile to handle potential NaN values if image is corrupted
            max_val = np.nanpercentile(channel, 99)
            min_val = np.nanpercentile(channel, 1)
            contrast = max_val - min_val
            # Ensure contrast is positive before log
            return contrast if contrast > 1e-6 else 1e-6

        con_r = calculate_contrast(img_r)
        con_g = calculate_contrast(img_g)
        con_b = calculate_contrast(img_b)

        log_con_r = math.log(con_r)
        log_con_g = math.log(con_g)
        log_con_b = math.log(con_b)

        # Check for non-finite results after log
        if not (np.isfinite(log_con_r) and np.isfinite(log_con_g) and np.isfinite(log_con_b)):
            # print("Warning: Non-finite log contrast in uiconm.")
            return 0

        return -(alpha_r * log_con_r + alpha_g * log_con_g + alpha_b * log_con_b)
    except Exception as e:
        # print(f"Error in uiconm: {e}")
        return 0 # Return 0 if any unexpected error occurs

def uism(img):
    try:
        # Check if image is grayscale, convert to color if so
        if len(img.shape) == 2 or img.shape[2] == 1:
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

        if img is None or img.size == 0: return 0

        img = img.astype(np.float32)/255.0
        if img.shape[2] != 3: return 0

        img_r = img[:,:,0]
        img_g = img[:,:,1]
        img_b = img[:,:,2]
        k1 = 5
        k2 = 5
        sym_r = eme(img_r, k1, k2)
        sym_g = eme(img_g, k1, k2)
        sym_b = eme(img_b, k1, k2)

        uism_val = (0.299 * sym_r + 0.587 * sym_g + 0.114 * sym_b)
        return uism_val if np.isfinite(uism_val) else 0 # Check final value
    except Exception as e:
        # print(f"Error in uism: {e}")
        return 0

def eme(img, k1, k2):
    try:
        # EME works on grayscale
        if len(img.shape) > 2: # If color, convert to gray
             if img.shape[2] == 3:
                  img = cv2.cvtColor((img * 255).astype(np.uint8), cv2.COLOR_BGR2GRAY)
             elif img.shape[2] == 1:
                  img = img.squeeze() # Remove channel dim if already gray but 3D
             else: return 0 # Unsupported channel number
        elif len(img.shape) != 2: return 0 # Not a 2D array

        img = img.astype(np.float32) # EME expects float but doesn't normalize internally

        m, n = img.shape
        # Ensure k1, k2 are valid
        k1 = max(1, min(k1, m))
        k2 = max(1, min(k2, n))

        # Calculate number of blocks, ensuring at least 1
        num_blocks_m = max(1, m // k1)
        num_blocks_n = max(1, n // k2)

        w = np.linspace(0, m - k1, num_blocks_m, dtype=int)
        h = np.linspace(0, n - k2, num_blocks_n, dtype=int)

        k = len(w) * len(h)
        if k == 0: return 0

        eme_val = 0.0
        processed_blocks = 0
        eps = 1e-9 # Small epsilon to prevent log(0) or division by zero

        for i in range(len(w)):
            for j in range(len(h)):
                block = img[w[i]:w[i] + k1, h[j]:h[j] + k2]
                if block.size == 0: continue

                max_val = np.max(block)
                min_val = np.min(block)

                # Ensure min_val is positive and max_val > min_val
                min_val_adjusted = max(min_val, eps) # Avoid min_val being 0 or negative
                max_val_adjusted = max(max_val, min_val_adjusted + eps) # Ensure max > min

                ratio = max_val_adjusted / min_val_adjusted

                try:
                    log_ratio = math.log(ratio)
                    if np.isfinite(log_ratio):
                        eme_val += log_ratio
                        processed_blocks += 1
                    # else: print(f"Warning: Non-finite log value in eme: ratio={ratio}")
                except ValueError:
                    # print(f"Warning: math domain error in log calculation: ratio={ratio}")
                    continue

        # Average over successfully processed blocks
        return (2.0 * eme_val) / processed_blocks if processed_blocks > 0 else 0
    except Exception as e:
        # print(f"Error in eme: {e}")
        return 0


def uicm(img):
    try:
        # Check if image is grayscale, convert to color if so
        if len(img.shape) == 2 or img.shape[2] == 1:
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

        if img is None or img.size == 0: return 0

        img = img.astype(np.float32)/255.0
        if img.shape[2] != 3: return 0

        img_r = img[:,:,0]
        img_g = img[:,:,1]
        img_b = img[:,:,2]

        RG = img_r - img_g
        YB = (img_r + img_g) / 2 - img_b

        # Use nanmean and nanstd to be robust to potential NaNs in image data
        rg_m = np.nanmean(np.abs(RG))
        yb_m = np.nanmean(np.abs(YB))
        rg_v = np.nanstd(RG)
        yb_v = np.nanstd(YB)

        # Check for NaN/Inf after calculation (just in case)
        if not (np.isfinite(rg_m) and np.isfinite(yb_m) and np.isfinite(rg_v) and np.isfinite(yb_v)):
             # print("Warning: NaN/Inf encountered in uicm calculation.")
             return 0

        # Original UIQM formula coefficients
        alpha = -0.0268
        beta = 0.1586

        # Add epsilon inside sqrt to prevent sqrt(0) -> NaN/Inf gradient issues if std=0
        mean_term = np.sqrt(rg_m*2 + yb_m*2 + 1e-10)
        std_term = np.sqrt(rg_v*2 + yb_v*2 + 1e-10)

        uicm_value = alpha * mean_term + beta * std_term
        return uicm_value if np.isfinite(uicm_value) else 0 # Final check
    except Exception as e:
        # print(f"Error in uicm: {e}")
        return 0

def uicqm(img):
    try:
        c1 = 0.0282 # Weight for UICM
        c2 = 0.2953 # Weight for UISM
        c3 = 3.5753 # Weight for UIConM

        uicm_val = uicm(img)
        uism_val = uism(img)
        uiconm_val = uiconm(img)

        # Check for NaN/Inf values from components before combining
        if not (np.isfinite(uicm_val) and np.isfinite(uism_val) and np.isfinite(uiconm_val)):
            # print("Warning: NaN or Inf encountered in UIQM component calculation.")
            return 0 # Return 0 if any component failed

        uiqm_score = c1 * uicm_val + c2 * uism_val + c3 * uiconm_val
        return uiqm_score if np.isfinite(uiqm_score) else 0 # Final check
    except Exception as e:
        # print(f"Error calculating final UIQM: {e}")
        return 0

# -------------------------

if __name__ == '__main__':
    import argparse
    parser = argparse.ArgumentParser()
    parser.add_argument('--results_path', type=str, default=None, help='Path for results')
    parser.add_argument('--ground_truth_path', type=str, default=None, help='Path for ground truth')
    args = parser.parse_args()

    corrected_path = args.results_path
    reference_path = args.ground_truth_path

    psnr_total = 0
    ssim_total = 0
    uicqm_total = 0
    num_psnr_ssim = 0 # Separate count for PSNR/SSIM
    num_uiqm = 0      # Separate count for UIQM in case it fails sometimes

    print(f"--- Starting Evaluation ---")
    print(f"Results Path: {corrected_path}")
    print(f"Ground Truth Path: {reference_path}")

    if corrected_path is None or reference_path is None:
        print("Error: Both --results_path and --ground_truth_path must be provided.")
        exit() # Exit if paths are missing

    if not os.path.isdir(corrected_path):
        print(f"Error: Results path not found or is not a directory: {corrected_path}")
        exit()
    if not os.path.isdir(reference_path):
         print(f"Error: Ground truth path not found or is not a directory: {reference_path}")
         exit()

    print(f"Calculating PSNR, SSIM, and UIQM...")
    # Get list of ground truth files (supports png, jpg, jpeg)
    image_patterns = [".png", ".jpg", "*.jpeg"]
    image_files = []
    for pattern in image_patterns:
        image_files.extend(glob.glob(os.path.join(reference_path, pattern)))

    print(f"Found {len(image_files)} images in ground truth path.")

    if not image_files:
        print(f"Error: No images found in ground truth path: {reference_path}")
    else:
        print(f"Iterating through {len(image_files)} ground truth files...")
        for idx, filename in enumerate(image_files):
            filename_base = os.path.basename(filename)
            corrected_file_path = os.path.join(corrected_path, filename_base)

            if not os.path.exists(corrected_file_path):
                # print(f"  -> Result file NOT found: {corrected_file_path}. Skipping.")
                continue # Skip this file if no exact match

            im_reference = cv2.imread(filename)
            im_corrected = cv2.imread(corrected_file_path)

            if im_reference is None:
                # print(f"  -> Error reading ground truth file: {filename}. Skipping.")
                continue
            if im_corrected is None:
                # print(f"  -> Error reading result file: {corrected_file_path}. Skipping.")
                continue

            # Ensure images have the same dimensions for PSNR/SSIM
            if im_reference.shape != im_corrected.shape:
                # print(f"  -> Warning: Shapes mismatch. GT={im_reference.shape}, Result={im_corrected.shape}. Resizing result.")
                try:
                    im_corrected_resized = cv2.resize(im_corrected, (im_reference.shape[1], im_reference.shape[0]), interpolation=cv2.INTER_CUBIC)
                except Exception as resize_e:
                    # print(f"  -> Error resizing image {filename_base}: {resize_e}. Skipping PSNR/SSIM.")
                    continue # Skip PSNR/SSIM if resize fails
            else:
                im_corrected_resized = im_corrected # Use original if shapes match

            # Calculate PSNR/SSIM using the potentially resized corrected image
            try:
                current_psnr = psnr(im_reference, im_corrected_resized, data_range=255)
                current_ssim = ssim(im_reference, im_corrected_resized, data_range=255, channel_axis=-1, win_size=min(7, im_reference.shape[0], im_reference.shape[1])) # Ensure win_size is odd and <= image dims
                if np.isfinite(current_psnr) and np.isfinite(current_ssim):
                    psnr_total += current_psnr
                    ssim_total += current_ssim
                    num_psnr_ssim += 1 # Increment only if both PSNR/SSIM are valid
                # else: print(f"  -> Warning: Non-finite PSNR/SSIM for {filename_base}.")
            except Exception as e:
                # print(f"  -> Error calculating PSNR/SSIM for {filename_base}: {e}. Skipping PSNR/SSIM.")
                pass


            # Calculate UIQM using the ORIGINAL corrected image (UIQM is no-reference)
            try:
                current_uiqm = uicqm(im_corrected) # Use original im_corrected
                if np.isfinite(current_uiqm): # Check if calculation resulted in NaN/Inf
                     uicqm_total += current_uiqm
                     num_uiqm += 1 # Increment UIQM counter only if successful
                # else: print(f"  -> Warning: UIQM calculation resulted in NaN/Inf for {filename_base}.")
            except Exception as e:
                # print(f"  -> Warning: Could not calculate UIQM for {filename_base}. Error: {e}. Skipping UIQM.")
                pass


        if num_psnr_ssim > 0 or num_uiqm > 0:
            print("\n--- Final Average Metrics ---")
            if num_psnr_ssim > 0:
                print(f"PSNR: {psnr_total / num_psnr_ssim:.4f}  (calculated on {num_psnr_ssim} images)")
                print(f"SSIM: {ssim_total / num_psnr_ssim:.4f}  (calculated on {num_psnr_ssim} images)")
            else:
                print("PSNR: Calculation failed for all images.")
                print("SSIM: Calculation failed for all images.")

            if num_uiqm > 0:
                 print(f"UIQM: {uicqm_total / num_uiqm:.4f}  (calculated on {num_uiqm} images)")
            else:
                 print("UIQM: Calculation failed for all images.")

            # print(f"Total ground truth images found: {len(image_files)}")
            # print(f"Matching result images found: {num_psnr_ssim if num_psnr_ssim > 0 else num_uiqm}") # Show how many pairs were actually processed
        else:
            print("No matching image pairs were found or processed successfully.")
```
